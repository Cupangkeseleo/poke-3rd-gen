```python
import os
from PIL import Image

dataset_dir = "gen3_dataset"
converted_count = 0
error_count = 0

print(f"Scanning {dataset_dir} to convert WEBP to JPG...")

for root, dirs, files in os.walk(dataset_dir):
    for file in files:
        file_path = os.path.join(root, file)
        
        try:
            with Image.open(file_path) as img:
                if img.format == 'WEBP':
                    rgb_img = img.convert('RGB')
                    rgb_img.save(file_path, 'JPEG')
                    print(f"Fixed: {file}")
                    converted_count += 1
                
        except Exception as e:
            print(f"Corrupt file found: {file_path} -> Deleting")
            os.remove(file_path)
            error_count += 1

print(f"\n--- DONE ---")
print(f"Fixed {converted_count} images.")
print(f"Deleted {error_count} unreadable files.")
```
